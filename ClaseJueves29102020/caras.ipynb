```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
rostro1 = cv2.imread('rostro1.png', 0)
```

```python
plt.imshow(rostro1, cmap='gray')
```

```python
familia = cv2.imread('familia.png', 0)
```

```python
plt.imshow(familia, cmap='gray')
```

```python
cascada = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detecta_cara(imagen):
    imagen1 = imagen.copy()
    rectangulos = cascada.detectMultiScale(imagen1)
    for(x,y,w,h) in rectangulos:
        cv2.rectangle(imagen1, (x,y), (x+w, y+h), (255,0, 0), 10)
        return imagen1
```

```python
resultado = detecta_cara(rostro1)
plt.imshow(resultado, cmap='gray')
```

```python
resultado = detecta_cara(familia)
plt.imshow(resultado, cmap='gray')
```
