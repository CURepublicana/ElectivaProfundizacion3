```python
#importamos lo que necesitamos del sistema
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('sudoku.jpg')
```

```python
plt.imshow(imagen, cmap='gray')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(imagen, cmap='gray')
```

```python
#detectar los datos verticales o sea x
sobelx = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=7)
plt.imshow(sobelx, cmap='gray')
```

```python
#detectar los datos verticales o sea x
sobely = cv2.Sobel(imagen, cv2.CV_64F, 0, 1, ksize=7)
plt.imshow(sobely, cmap='gray')
```

```python
#mezclar imagenes
imagen2 = cv2.addWeighted(src1=sobelx,alpha=0.5, src2=sobely, beta=0.5, gamma=0)
```

```python
plt.imshow(imagen2, cmap='gray')
```

```python
#Laplacian - mtodo para obtener los datos 
imagen3 = cv2.Laplacian(imagen, cv2.CV_64F)
plt.imshow(imagen3, cmap='gray')
```
