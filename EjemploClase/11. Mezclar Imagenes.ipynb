```python
import matplotlib.pyplot as plt
import cv2
```

```python
imagen1 = cv2.imread('perro.jpg')
```

```python
imagen1 = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen1)
```

```python
imagen2 = cv2.imread('Marca.jpg')
imagen2 = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
imagen1.shape
```

```python
imagen2.shape
```

```python
imagen1 = cv2.resize(imagen1, (1200,1200))
plt.imshow(imagen1)
```

```python
imagen2 = cv2.resize(imagen2, (1200,1200))
plt.imshow(imagen2)
```

```python
imagen_mezcla = cv2.addWeighted(src1=imagen1, alpha=0.8, src2=imagen2, beta=0.3, gamma=0)
```

```python
plt.imshow(imagen_mezcla)
```
