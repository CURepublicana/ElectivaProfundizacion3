```python
import matplotlib.pyplot as plt
import cv2
```

```python
imagen1 = cv2.imread('perro.jpg')
imagen1 = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)
plt.imshow(imagen1)
```

```python
imagen2 = cv2.imread('perro.jpg', 0)
plt.imshow(imagen2, cmap='gray')
```

```python
mitad, imagen2 =cv2.threshold(imagen2, 255/2, 255, cv2.THRESH_BINARY)
plt.imshow(imagen2, cmap='gray')
```

```python
mitad, imagen2 =cv2.threshold(imagen2, 255/2, 255, cv2.THRESH_BINARY_INV)
plt.imshow(imagen2, cmap='gray')
```
