```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('perro.jpg')
```

```python
plt.imshow(imagen)
```

```python
imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
imagen3 = cv2.cvtColor(imagen, cv2.COLOR_RGB2GRAY)
plt.imshow(imagen3)
```

```python
imagen3 = cv2.cvtColor(imagen, cv2.COLOR_LRGB2LUV)
plt.imshow(imagen3)
```

```python
imagen4 = cv2.cvtColor(imagen, cv2.COLORMAP_HOT)
plt.imshow(imagen4)
```

```python
imagen5 = cv2.cvtColor(imagen, cv2.COLOR_RGB2HLS_FULL)
plt.imshow(imagen5)
```
